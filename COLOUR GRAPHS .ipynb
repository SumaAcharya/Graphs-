```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,10,11)
y=2*x**3
plt.figure(figsize=(5,3), dpi=100)

plt.plot(x,y , lw=1, color="red", linestyle="-", marker="o")
plt.title("y=f(x)", fontsize="30")
plt.xlabel(('x'),size="20")
plt.ylabel(('y'),size="20")
plt.xlim(0,10)
plt.ylim(0,2000)
plt.grid()
plt.xlim(x[0],x[-1])
plt.ylim(y[0],y[-1])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,20,21)
y1=2*x**2
y2=x**2-3*x

myfig=plt.figure()
ax1=myfig.add_axes([0,0,1,1])
ax2=myfig.add_axes([0.1,0.3,0.4,0.4])
ax3=myfig.add_axes([0.6,0.6,0.3,0.3])
ax3.plot(x,y2,color='green')
ax1.plot(x,y1)
ax2.plot(x,y2, color='red')
ax2.set_title("y2")
ax3.set_title("y3")


plt.show()
```

```python
import numpy as py
import matplotlib.pyplot as plt

theta=np.linspace(0,360,30)
y1=np.sin(theta*np.pi/180)
y2=np.cos(theta*np.pi/180)

x=np.linspace(1,10,11)
z1=x**3
z2=x**2

plt.figure(figsize=(10,10))
plt.subplot(221)
plt.plot(theta,y1)
plt.title('sin'r'($\theta{}$)')
plt.tight_layout(pad=5)
plt.subplot(222)
plt.plot(theta,y2, color='orange')
plt.title('cos'r'($\theta{}$)')
plt.subplot(223)
plt.plot(x,z1, color='red')
plt.title('z='r'$x^{3}$')
plt.tight_layout(pad=5)
plt.subplot(224)
plt.plot(x,z2, color='violet')
plt.title('z='r'$x^{2}$')



plt.show()
        
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gsp

x=np.linspace(0,5,250)
y1=2*x
y2=(3*x**2)-np.exp(-(x-0.2)**2)
y3=x**3-x**2-1

fig= plt.figure(figsize=(10,10), facecolor=(0.5,0.5,0.5))
gs=gsp.GridSpec(2,2)

ax1= fig.add_subplot(gs[0,1],)
ax2= fig.add_subplot(gs[1,0],)
ax3= fig.add_subplot(gs[1,1])

ax1.plot(x,y1, color='red')
ax1.set_title('2x')
ax2.plot(x,y2,color='orange')
ax2.set_title('$3x^2-e^{-(x-0.2)^2}$')
ax3.plot(x,y3)
ax3.set_title('$x^3-x^2-1$')

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

xp=np.linspace(-5,5,5)
yp=np.linspace(-5,5,5)

X,Y=np.meshgrid(xp,yp)

z=np.sin(X)+np.cos(Y)
plt.contourf(X,Y,z,10,cmap='jet')
plt.colorbar()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

xp=np.linspace(-5,5,5)
yp=np.linspace(-5,5,5)

X,Y=np.meshgrid(xp,yp)

z=np.sqrt(X**2+Y**2)
plt.contourf(X,Y,z,50,cmap='plasma')
plt.colorbar()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

x=np.arange(-5,5,0.1)
y=np.arange(-5,5,0.1)
X,Y= np.meshgrid(x,y)
Z=(X**2+Y**2)

plt.figure(figsize=(5,5))
ax=plt.axes(projection='3d')
mysurf=ax .plot_surface(X,Y,Z,cmap='jet')
plt.colorbar(mysurf)
cset=plt.contourf(X,Y,Z ,zdir='z',offset=0,cmap='hot')
ax.set_xlabel('x')
ax.set_ylabel('y')

plt.show()
```

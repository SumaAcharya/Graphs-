# 3D Graph plot

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

x=np.arange(-5,5,0.1)
y=np.arange(-5,5,0.1)
X,Y= np.meshgrid(x,y)
Z=(X**2+Y**2)

plt.figure(figsize=(5,5))
ax=plt.axes(projection='3d')
mysurf=ax .plot_surface(X,Y,Z,cmap='jet')
plt.colorbar(mysurf)
cset=plt.contourf(X,Y,Z ,zdir='z',offset=0,cmap='hot')
ax.set_xlabel('x')
ax.set_ylabel('y')

plt.show()
```

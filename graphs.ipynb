```python
from numpy import pi,sin,cos,sqrt
import matplotlib.pyplot as plt
import numpy as np
plt.axes(projection='polar')

a=7
b=7
theta=np.arange(0,2*pi,pi/180)
for t in theta:
    r=(a*b)/sqrt((a*sin(t))**2+(b*cos(t))**2)
    plt.polar(t,r,'r')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi,sin,cos

u0= 40
g=9.81

theta1=45
theta2=30
ux1= u0*cos(theta1*pi/180)
uy1= u0*sin(theta1*pi/180)

ux2= u0*cos(theta2*pi/180)
uy2= u0*sin(theta2*pi/180)

t_total_1=2*uy1/g
t_total_2=2*uy2/g
t1=np.linspace(0,t_total_1,100)
t2=np.linspace(0,t_total_2,100)

sx1=ux1*t1
sy1=(uy1*t1)-(0.5*g*t1**2)

sx2=ux2*t2
sy2=(uy2*t2)-(0.5*g*t2**2)

plt.figure(figsize=(10,5))
plt.plot(sx1,sy1,label='theta1=45')

plt.plot(sx2,sy2,label='theta2=30')
plt.ylim(0,50)
plt.legend()
plt.show()
```

# Shearing Force and Bending Moment

```python
import numpy as np
import matplotlib.pyplot as plt

# input
L=20 #length
w=50  # work

#reaction
R=L*w/2
x=np.linspace(0,L,200)

X=[]
SF=[]
M=[]

for l in x:
    sf= R-(w*l) #shearing force
    m = (R*l)-(w*l**2/2) # moment 
    X.append(l)
    SF.append(sf)
    M.append(m)
    
plt.subplot(2,1,1)
plt.plot(X,SF, color='purple')
plt.fill_between(X,SF, color='purple',alpha=0.5, hatch="||")
plt.title("Shearing force diagram")
plt.ylabel("shearing force(SFD)")
plt.tight_layout(pad=4.0)

plt.subplot(2,1,2)
plt.fill_between(X,M,color='blue', alpha=0.5, hatch="||")
plt.title("Bending Moment Diagram")
plt.ylabel("Bending Moment(BMD)")
plt.plot(X,M,color='blue')
plt.show()
```

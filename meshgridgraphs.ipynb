# Examples of meshgrid contouring colour graphs.

```python
import numpy as np
import matplotlib.pyplot as plt

xp=np.linspace(-5,5,5)
yp=np.linspace(-5,5,5)

X,Y=np.meshgrid(xp,yp)

z=np.sin(X)+np.cos(Y)
plt.contourf(X,Y,z,10,cmap='jet')
plt.colorbar()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

xp=np.linspace(-5,5,5)
yp=np.linspace(-5,5,5)

X,Y=np.meshgrid(xp,yp)

z=np.sqrt(X**2+Y**2)
plt.contourf(X,Y,z,50,cmap='plasma')
plt.colorbar()
plt.show()
```
